```python
# perform labeled encoding
# convert categorical coloumn values to corresponding numerical values using labeled encoding
from sklearn.preprocessing import LabelEncoder
l_encoder=LabelEncoder()
col_list=['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']
for i in range(len(col_list)):
    mushroom[col_list[i]]=l_encoder.fit_transform(mushroom[col_list[i]])
mushroom    
    
```

```python
# performing one hot encoding
mushroom1=pd.read_csv('mushrooms.csv')
mushroom1
```

```python
mushroom1=pd.get_dummies(mushroom1,columns=['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'])
```

```python
mushroom1
```

```python
# split the dataset into two as indepentend and depentend coloumns x and y
x=mushroom.drop('class',axis=1)
x
```

```python
y=mushroom['class']
y
```

```python
# performing mechine learning algorithm
# split again to train and train dataset for model training and test process
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=40,test_size=0.2)
```

```python
# 1) logistic regression
from sklearn.linear_model import LogisticRegression
model0=LogisticRegression()
model0.fit(x_train,y_train)
y_predict0=model0.predict(x_test)
y_predict0
```

```python
from sklearn.metrics import accuracy_score
a0=accuracy_score(y_test,y_predict0)
print('accuracy score (logistic regression):',a0)
```

```python
# 2) performing KNN
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
x=[5,7,9,11,13]
y=[]
for i in x:
    model1=KNeighborsClassifier(n_neighbors=i)
    model1.fit(x_train,y_train)
    y_predict=model1.predict(x_test)
    a=accuracy_score(y_test,y_predict)
    y.append(a)
plt.plot(x,y,'*-')
    
    
```

```python
# 3) performing dicession tree
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
y_predict2=model2.predict(x_test)
a2=accuracy_score(y_test,y_predict2)
print("accuracy score(DecisionTreeClassifier):",a2)
```

```python
# 4) performing random forest algorithm
from sklearn.ensemble import RandomForestClassifier 
model3=RandomForestClassifier(n_estimators=8)
model3.fit(x_train,y_train)
y_predict3=model3.predict(x_test)
a3=accuracy_score(y_test,y_predict3)
print("Accuracy score(Random forest)",a3)
```

```python
# 5) performing SVM
from sklearn.svm import SVC
model4=SVC()
model4.fit(x_train,y_train)
y_predict4=model4.predict(x_test)
a4=accuracy_score(y_test,y_predict4)
print("Accuracy score(SVM)",a4)
```
